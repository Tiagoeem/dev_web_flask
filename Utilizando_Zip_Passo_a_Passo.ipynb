Vamos simular a respota de um registro, o valor vem no formato de uma tupla com N posições.

```python
linha = ('Silva', 'Fulano', '123.123.123-12', 1)
print(linha)
```

```python
colunas_clientes = ['sobrenome', 'nome', 'cpf', 'id_cliente']
print(colunas_clientes)
```

A operação zip, cria pares entre cada valor dos elementos passados como parametro, veja o output a seguir.

Ótimo tutorial: https://realpython.com/python-zip-function/

Implementação interna na documentação: https://docs.python.org/3.3/library/functions.html

image.png

```python
# Utilizando uma lista para que possamos visualizar os pares formados
apenas_para_visualizar_os_pares_list = list(zip(colunas_clientes, linha))
print(apenas_para_visualizar_os_pares_list)
```

Finalmente, transformando o zip() em dicionário.

```python
resp_dict = dict(zip(colunas_clientes, linha))
print(resp_dict)
```
